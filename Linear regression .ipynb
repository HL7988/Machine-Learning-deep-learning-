```python
import numpy as np
import random
# let m= training example, n= number of features 
# dims of y=(1,m), X =(n,m), w(n,1)
class LinearReg():
    def __init__(self):
        self.learning_rate=0.01
        self.total_iterations=10000
        
    def y_hat(self,x,w):
        return np.dot(w.T,x)
    
    def loss(self,yhat, y):# yhat= w1x1+ w2x2 +.....
        L=  (1/self.m)* np.sum(np.power(yhat-y, 2))
        #L=np.sum((y - yhat) ** 2) / len(y)
        return L
    
    def gradient_descent(self, w, x, y, yhat):
        #shape y =(1,m), x is (n,m)
        # output should be (n,1)
        
        dldw= 2/self.m * np.dot(x, (yhat-y).T)
        w=w - self.learning_rate*dldw
        return w
         
    def main(self,x,y):
        # here we have also considered x1 for all training examples so we need to
        #add 1 to all training example i.e we have to add one more input vector all columns 
        
        
        x1=np.ones((1, x.shape[1]))
        x= np.append(x, x1, axis=0)
        
        self.m= x.shape[1]
        self.n= x.shape[0]
        
        w=np.zeros((self.n, 1))
        
        for i in range(self.total_iterations +1):
            
    
            yhat= self.y_hat(x,w)
            loss= self.loss(yhat,y)

            if i%100 ==0:
                print(f'cost of iterations {i} is {loss}')
            w=self.gradient_descent(w,x,y,yhat)
        return w
if __name__=='__main__':
    x=np.random.randn(1,500)
   
    y=3*x + np.random.randn(1,500)*0.1
    regression= LinearReg()
    w= regression.main(x,y)
```
