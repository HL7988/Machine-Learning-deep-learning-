```python
import cv2 as cv
import numpy as np
import os 
import pandas as pd
```

```python
df=pd.read_csv('Downloads/08-11-2019-2-2_155.csv')
```

```python
frames=df.iloc[:,0]
```

```python
frames=frames.to_numpy()
```

```python
frames 
```

```python
vidcap = cv.VideoCapture('Downloads/08-11-2019-2-2.MP4')
success,image = vidcap.read()
print(success)

i=0 #npframe iterator
cap=0 #vidcapture iterator

while success:
    if cap==frames[i]:
        cv.imwrite("./dataa1/frame%d.jpg" % frames[i], image)     # save frame as JPEG file      
        print('Read a new frame: ', success)
        i=i+1
    cap=cap+1
    success,image = vidcap.read()
    if i==707:
        break
```
